```python
#Sayısal değikenlerin Dağılımını ifade etmek için kullanılan Veri görselleiitirme tekniği
```

```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
sns.distplot(df.price,kde = False,bins = None,color = "R"); # Histogram  # bins grafiği hassaslaştırır. eğer 10 sa 10 sütun olarak iekilde çizer
#kde = True da ise yopunluk ekler,#Olasılık yoğunluk grafiğine göre çizer
```

```python
?sns.distplot
```

```python
sns.distplot(df.price,kde = True,bins = None,color = "R",hist=False ) # Sadece yoğunluk 
```

```python
#Sayısal durumda her ikiside kullanılabilir. Özellikler bir kullanım alanı yok
```

```python
sns.kdeplot(df.price,shade=True ); # Sadece yoğunluk 
```

```python
#Histogram ve Yoğunluk Çarprazlaması
```

```python
#Amaç Grafiğin gizlediği şeyleri anlatır. Mesela yukarda mavi kısmı oluşturan nedir.
```

```python
(sns
 .FacetGrid(df,
              hue = "cut",
              height=6,
              xlim=(0,15000))# Hue Boyut Ekleme Argümanı
.map(sns.kdeplot,"price",shade = True)
.add_legend()
);
```

```python
sns.catplot(x = "cut",y = "price",hue= "color",kind = "point",data = df);
```
