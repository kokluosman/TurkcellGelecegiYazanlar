```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
#Total_bill = Yemeğin Toplam Fiyatı
#tip = bahşiş
#sex = cinsiyet 0-male 1-female
#smoker = sigara içen varmı 0 yok 1 var
#day = gün
#time= zaman (0 = day 1 = night)
#size= gruptaki kişi sayısı
```

```python
#Korelasyon değişkenler arasındaki ilişkiyi ifade eder.
```

```python
#Saçılım Grafiği Yaklaşımı scatterplot ######Kategorik ise sütun borplot,Sayısal ise histogram,catplot
```

```python
#İki değişkenli veri görselleştirme
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df);
```

```python
#Çaprazlama
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="time",data=df);
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="time",data=df,style="time");
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="day",data=df,style="time");#4 özellik incelemiş oluyoruz
```

```python
sns.scatterplot(x="total_bill",y="tip",size="size",hue="day",data=df);#4 özellik incelemiş oluyoruz
```

```python
#Doğrusal İlişkinin Gösterilmesi
```

```python
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x="total_bill",y="tip",data=df);#Lineermodelin kısaltması
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",data=df);#Lineermodelin kısaltması
```

```python
df.groupby(["smoker","sex"])["tip"].describe().T
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",data=df);#col iki farklı grafiğe dönüştürdü
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",row="sex",data=df);#col iki farklı grafiğe dönüştürdü
```
