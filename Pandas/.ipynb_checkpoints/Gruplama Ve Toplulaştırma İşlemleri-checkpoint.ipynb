```python
#Aggregate/Filter/Transform/Apply
```

```python
df = pd.DataFrame({"Gruplar": ["a","b","c","a","b","c"],
                  "Degisken1":[10,23,33,22,11,99],
                  "Degisken2":[100,253,333,262,111,969]},
                 columns=["Gruplar","Degisken1","Degisken2"])
df
```

```python
#Aggregate
```

```python
df.groupby("Gruplar").mean()
```

```python
df.groupby("Gruplar").aggregate([min,np.median,max]) # aggregate bize istediğimiz toplulaştırıcı fonksiyonları bir arada kullanmayı sağlar
```

```python
df.groupby("Gruplar").aggregate({"Degisken1": "min","Degisken2":max})
```

```python
#Filter Burada tanımladığım fonksiyonu bunun üzerinde çalıştırıyorum
```

```python
df
```

```python
def filter_func(x):
    return x["Degisken1"].std()>9
```

```python
df.groupby("Gruplar").filter(filter_func) #Standart sapması 9'dan büyük olanları getirdi
```

```python
#Transform # Burada Değişkenler üzerinde dönüştürmeyi kendin tanımlayıp
# Filterdakine benzer şekilde dataset üzerinde gezdirebiliriz.
```

```python
df
```

```python
df_a = df.iloc[:,1:3]
```

```python
df_a.transform(lambda x: (x-x.mean())/(x.std()))
```

```python
#Pivot Tablolar 
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# Cinsiyete göre gruplayıp sağ kalma durumlarını ifade eden survived inceleyelim 
```

```python
titanic.groupby("sex")["survived"].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean")
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
#pivot ile pivot table
```

```python
titanic.pivot_table("survived",index="sex",columns="class")
```

```python
titanic.age.head()
```

```python
age = pd.cut(titanic["age"],[0,18,90])
age.head()
```

```python
titanic.pivot_table("survived",["sex",age],"class")
```
